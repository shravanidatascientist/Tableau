### Step 3 — Data Cleaning & Transformation

### Replace experience level codes

```python

dataset['experience_level'] = dataset['experience_level'].replace({
    'EN': 'Entry Level',
    'MI': 'Mid Level',
    'SE': 'Senior Level',
    'EX': 'Executive'
})
```

```python
dataset
```

### Replace employment type codes


```python
dataset['employment_type'] = dataset['employment_type'].replace({
    'FT': 'Full-Time',
    'PT': 'Part-Time',
    'CT': 'Contract',
    'FL': 'Freelance'
})
```

```python
dataset
```

###  Replace remote ratio numbers with labels

```python
dataset['remote_ratio'] = dataset['remote_ratio'].replace({
    0: 'On-Site',
    50: 'Hybrid',
    100: 'Remote'
})
```

```python
dataset
```

### Replace company size codes

```python
dataset['company_size'] = dataset['company_size'].replace({
    'S': 'Small',
    'M': 'Medium',
    'L': 'Large'
})
```

```python
dataset
```

#### Drop raw salary columns (we only need salary_in_usd)

```python
print(dataset.columns.tolist())
```

#### Rename columns to clean readable names for Tableau

```python
dataset = dataset.rename(columns={
    'work_year': 'Work Year',
    'experience_level': 'Experience Level',
    'employment_type': 'Employment Type',
    'job_title': 'Job Title',
    'salary_in_usd': 'Salary (USD)',
    'employee_residence': 'Employee Country',
    'remote_ratio': 'Work Mode',
    'company_location': 'Company Country',
    'company_size': 'Company Size'
})

dataset
```

### Step 4 — Exploratory Data Analysis (EDA)

#### Remove duplicates & outliers

```python
# Remove duplicate rows

before = len(dataset)

dataset = dataset.drop_duplicates()
after = len(dataset)
print(f'Duplicates removed: {before - after} rows')
print(f'Rows remaining: {after}')
```

#### Remove extreme salary outliers (keep 5th to 99th percentile)

```python
low = dataset['Salary (USD)'].quantile(0.05)
high = dataset['Salary (USD)'].quantile(0.99)
dataset = dataset[(dataset['Salary (USD)'] >= low) & (dataset['Salary (USD)'] <= high)]
print(f'After removing outliers: {len(dataset)} rows')
print(f'Salary range: ${low:,.0f} to ${high:,.0f}')
```

```python
print('================== TOP 15 JOB TITLES ==========================')

print(dataset['Job Title'].value_counts().head(15))

print('\n=================== AVG SALARY BY EXPERIENCE =================')

print(dataset.groupby('Experience Level')['Salary (USD)'].mean()
        .sort_values(ascending=False)
        .apply(lambda x: f'${x:,.0f}'))

print('\n================== TOP 10 HIRING COUNTRIES =====================')

print(dataset['Company Country'].value_counts().head(10))

print('\n====================WORK MODE SPLIT =============================')

print(dataset['Work Mode'].value_counts())
```
